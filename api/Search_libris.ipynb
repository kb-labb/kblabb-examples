# Find people in Libris
This is a simple script to search Libris for person names through Python. 

```python
import requests
import json
import re
```

```python
headers = {"Accept": "application/ld+json"}
```

```python
raw_name = "Thomas av Aquino"

if re.match(r'\S*\s\S*$', raw_name):
    given_name, family_name = raw_name.lower().split()
    r = requests.get(
        f"https://libris-qa.kb.se/find.jsonld?givenName={given_name}&familyName={family_name}&@type=Person",
        headers=headers
    )
elif re.match(r'\S*$', raw_name):
    r = requests.get(
        f"https://libris-qa.kb.se/find.jsonld?name={raw_name.lower()}&@type=Person",
        headers=headers
    )
else:
    r = requests.get(
        f"https://libris-qa.kb.se/find.jsonld?q={raw_name.lower()}&@type=Person",
        headers=headers
    )
    
```

```python
data = json.loads(r.content)
```

```python
data
```

```python
id_list = []
for item in data["items"]:
    id_list.append(item["@id"])
```
